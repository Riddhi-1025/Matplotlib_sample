```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
ypoints = np.array([10,20,30])
plt.plot(ypoints, linewidth = 5)
plt.show()
```

```python
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,5,11])

plt.plot(y1,color="red")
plt.plot(y2,color="black")

plt.show()
```

```python
plt.title("sample")
plt.grid(color="green",linestyle="dotted")
```

```python
x=np.array(['A','B','C','D'])
y=np.array([60,35,18,32])
plt.barh(x,y,color="black")
plt.show()
```

```python
y=np.array([35,25,25,15])
mylabels = ["Apples", "Bananas", "Cherries","grapes"]

plt.pie(y)
plt.legend(labels=mylabels)
plt.show()
```

```python
y=np.array([50,60,30,20])
z=np.array([23,180,900,800])
plt.plot(y,z)
plt.show()
```

```python
x=np.array(['x','y','z','c'])
plt.hist(x)
plt.show()
```
